```python
# EDA COMPLETO - LUGARES TUR√çSTICOS Y CULTURALES DE LIMA
# =========================================================
# INCLUYE AN√ÅLISIS DETALLADO DE REVIEWS

# 1. INSTALACI√ìN Y CONFIGURACI√ìN
# =========================================================
# Instalar librer√≠as necesarias para an√°lisis de texto
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Para an√°lisis de texto
from textblob import TextBlob
from wordcloud import WordCloud
import re
from langdetect import detect, LangDetectException

# Configuraci√≥n de visualizaci√≥n
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 10

# Cargar datos
# df = pd.read_csv('tu_archivo.csv')

print("=" * 80)
print("AN√ÅLISIS EXPLORATORIO DE DATOS - LUGARES DE LIMA")
print("=" * 80)

# 2. INFORMACI√ìN GENERAL DEL DATASET
# =========================================================
print("\nüìä INFORMACI√ìN GENERAL DEL DATASET")
print("-" * 80)
print(f"Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
print(f"\nColumnas: {list(df.columns)}")

# 3. PROCESAMIENTO DE REVIEWS
# =========================================================
print("\n\nüìù PROCESAMIENTO DE REVIEWS")
print("-" * 80)

# Funci√≥n para extraer informaci√≥n de reviews
def extract_review_info(reviews_str):
    """Extrae informaci√≥n detallada de las reviews"""
    if pd.isna(reviews_str) or reviews_str == '[]':
        return []
    try:
        if isinstance(reviews_str, str):
            reviews = eval(reviews_str)
        else:
            reviews = reviews_str
        return reviews
    except:
        return []

# Extraer todas las reviews individuales
all_reviews = []
for idx, row in df.iterrows():
    reviews = extract_review_info(row['reviews'])
    for review in reviews:
        review_data = {
            'place_name': row['name'],
            'place_rating': row['rating'],
            'categoria': row['categoria'],
            'distrito': row['distrito'],
            'author': review.get('author', ''),
            'rating': review.get('rating', None),
            'text': review.get('text', ''),
            'time': review.get('time', '')
        }
        all_reviews.append(review_data)

reviews_df = pd.DataFrame(all_reviews)
print(f"Total de reviews extra√≠das: {len(reviews_df):,}")
print(f"Lugares con al menos una review: {reviews_df['place_name'].nunique()}")
print(f"Promedio de reviews por lugar: {len(reviews_df) / df.shape[0]:.2f}")

# 4. ESTAD√çSTICAS B√ÅSICAS DE REVIEWS
# =========================================================
print("\n\nüìä ESTAD√çSTICAS B√ÅSICAS DE REVIEWS")
print("-" * 80)

# Reviews por lugar
df['num_reviews'] = df['reviews'].apply(lambda x: len(extract_review_info(x)))
print(f"Lugar con m√°s reviews: {df.loc[df['num_reviews'].idxmax(), 'name']} ({df['num_reviews'].max()} reviews)")
print(f"Promedio de reviews por lugar: {df['num_reviews'].mean():.2f}")
print(f"Mediana de reviews por lugar: {df['num_reviews'].median():.0f}")

# Distribuci√≥n de ratings en reviews
print(f"\nDistribuci√≥n de ratings en reviews:")
print(reviews_df['rating'].value_counts().sort_index())
print(f"\nRating promedio en reviews: {reviews_df['rating'].mean():.2f}")

# Reviews con y sin texto
reviews_df['has_text'] = reviews_df['text'].apply(lambda x: len(str(x).strip()) > 0)
print(f"\nReviews con texto: {reviews_df['has_text'].sum()} ({reviews_df['has_text'].sum()/len(reviews_df)*100:.1f}%)")
print(f"Reviews sin texto: {(~reviews_df['has_text']).sum()} ({(~reviews_df['has_text']).sum()/len(reviews_df)*100:.1f}%)")

# Visualizaci√≥n
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Distribuci√≥n de n√∫mero de reviews por lugar
axes[0, 0].hist(df['num_reviews'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('N√∫mero de Reviews')
axes[0, 0].set_ylabel('Cantidad de Lugares')
axes[0, 0].set_title('Distribuci√≥n del N√∫mero de Reviews por Lugar', fontweight='bold')
axes[0, 0].axvline(df['num_reviews'].mean(), color='red', linestyle='--', linewidth=2, label=f'Media: {df["num_reviews"].mean():.1f}')
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)

# Distribuci√≥n de ratings en reviews
rating_counts = reviews_df['rating'].value_counts().sort_index()
axes[0, 1].bar(rating_counts.index, rating_counts.values, color=['#d32f2f', '#f57c00', '#fbc02d', '#7cb342', '#388e3c'],
               edgecolor='black', alpha=0.8)
axes[0, 1].set_xlabel('Rating')
axes[0, 1].set_ylabel('Cantidad de Reviews')
axes[0, 1].set_title('Distribuci√≥n de Ratings en Reviews Individuales', fontweight='bold')
axes[0, 1].grid(axis='y', alpha=0.3)
for i, v in enumerate(rating_counts.values):
    axes[0, 1].text(rating_counts.index[i], v + 50, str(v), ha='center', fontweight='bold')

# Top 15 lugares con m√°s reviews
top_reviewed = df.nlargest(15, 'num_reviews')[['name', 'num_reviews']].sort_values('num_reviews')
axes[1, 0].barh(range(len(top_reviewed)), top_reviewed['num_reviews'], color='coral')
axes[1, 0].set_yticks(range(len(top_reviewed)))
axes[1, 0].set_yticklabels(top_reviewed['name'], fontsize=9)
axes[1, 0].set_xlabel('N√∫mero de Reviews')
axes[1, 0].set_title('Top 15 Lugares con M√°s Reviews', fontweight='bold')
for i, v in enumerate(top_reviewed['num_reviews'].values):
    axes[1, 0].text(v + 1, i, str(v), va='center')

# Reviews por categor√≠a
cat_reviews = reviews_df.groupby('categoria').size().sort_values(ascending=False).head(15)
axes[1, 1].barh(range(len(cat_reviews)), cat_reviews.values, color='lightgreen')
axes[1, 1].set_yticks(range(len(cat_reviews)))
axes[1, 1].set_yticklabels(cat_reviews.index, fontsize=9)
axes[1, 1].set_xlabel('N√∫mero de Reviews')
axes[1, 1].set_title('Top 15 Categor√≠as con M√°s Reviews', fontweight='bold')
axes[1, 1].invert_yaxis()
for i, v in enumerate(cat_reviews.values):
    axes[1, 1].text(v + 10, i, str(v), va='center')

plt.tight_layout()
plt.show()

# 5. AN√ÅLISIS DE TEXTO DE REVIEWS
# =========================================================
print("\n\nüìñ AN√ÅLISIS DE TEXTO DE REVIEWS")
print("-" * 80)

# Filtrar reviews con texto
reviews_with_text = reviews_df[reviews_df['has_text']].copy()
print(f"Analizando {len(reviews_with_text):,} reviews con texto...")

# Calcular longitud de texto
reviews_with_text['text_length'] = reviews_with_text['text'].apply(lambda x: len(str(x)))
reviews_with_text['word_count'] = reviews_with_text['text'].apply(lambda x: len(str(x).split()))

print(f"\nEstad√≠sticas de longitud de texto:")
print(f"  - Longitud promedio: {reviews_with_text['text_length'].mean():.0f} caracteres")
print(f"  - Palabras promedio: {reviews_with_text['word_count'].mean():.1f} palabras")
print(f"  - Review m√°s larga: {reviews_with_text['text_length'].max()} caracteres")
print(f"  - Review m√°s corta: {reviews_with_text['text_length'].min()} caracteres")

# Detectar idioma
def detect_language(text):
    try:
        return detect(str(text))
    except:
        return 'unknown'

reviews_with_text['language'] = reviews_with_text['text'].apply(detect_language)
print(f"\nDistribuci√≥n de idiomas:")
print(reviews_with_text['language'].value_counts())

# Visualizaci√≥n
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Distribuci√≥n de longitud de texto
axes[0, 0].hist(reviews_with_text['text_length'], bins=50, color='mediumpurple', edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Longitud del Texto (caracteres)')
axes[0, 0].set_ylabel('Frecuencia')
axes[0, 0].set_title('Distribuci√≥n de Longitud de Reviews', fontweight='bold')
axes[0, 0].axvline(reviews_with_text['text_length'].mean(), color='red', linestyle='--', linewidth=2,
                   label=f'Media: {reviews_with_text["text_length"].mean():.0f}')
axes[0, 0].legend()

# Distribuci√≥n de palabras
axes[0, 1].hist(reviews_with_text['word_count'], bins=50, color='lightcoral', edgecolor='black', alpha=0.7)
axes[0, 1].set_xlabel('N√∫mero de Palabras')
axes[0, 1].set_ylabel('Frecuencia')
axes[0, 1].set_title('Distribuci√≥n de Cantidad de Palabras', fontweight='bold')
axes[0, 1].axvline(reviews_with_text['word_count'].mean(), color='red', linestyle='--', linewidth=2,
                   label=f'Media: {reviews_with_text["word_count"].mean():.1f}')
axes[0, 1].legend()

# Idiomas
lang_counts = reviews_with_text['language'].value_counts().head(10)
axes[1, 0].bar(range(len(lang_counts)), lang_counts.values, color='teal', alpha=0.7, edgecolor='black')
axes[1, 0].set_xticks(range(len(lang_counts)))
axes[1, 0].set_xticklabels(lang_counts.index, rotation=45)
axes[1, 0].set_ylabel('Cantidad de Reviews')
axes[1, 0].set_title('Distribuci√≥n de Idiomas en Reviews', fontweight='bold')
axes[1, 0].grid(axis='y', alpha=0.3)

# Relaci√≥n longitud vs rating
axes[1, 1].scatter(reviews_with_text['word_count'], reviews_with_text['rating'],
                   alpha=0.5, color='orange', edgecolors='black', linewidth=0.5, s=30)
axes[1, 1].set_xlabel('N√∫mero de Palabras')
axes[1, 1].set_ylabel('Rating')
axes[1, 1].set_title('Relaci√≥n: Longitud de Review vs Rating', fontweight='bold')
axes[1, 1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 6. AN√ÅLISIS DE SENTIMIENTO
# =========================================================
print("\n\nüòä AN√ÅLISIS DE SENTIMIENTO")
print("-" * 80)

# Funci√≥n para an√°lisis de sentimiento
def analyze_sentiment(text):
    try:
        analysis = TextBlob(str(text))
        # Polarity: -1 (negativo) a 1 (positivo)
        polarity = analysis.sentiment.polarity
        # Subjectivity: 0 (objetivo) a 1 (subjetivo)
        subjectivity = analysis.sentiment.subjectivity

        if polarity > 0.1:
            sentiment = 'Positivo'
        elif polarity < -0.1:
            sentiment = 'Negativo'
        else:
            sentiment = 'Neutral'

        return polarity, subjectivity, sentiment
    except:
        return 0, 0, 'Neutral'

reviews_with_text[['polarity', 'subjectivity', 'sentiment']] = reviews_with_text['text'].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)

print(f"Distribuci√≥n de sentimientos:")
print(reviews_with_text['sentiment'].value_counts())
print(f"\nPolarity promedio: {reviews_with_text['polarity'].mean():.3f}")
print(f"Subjectivity promedio: {reviews_with_text['subjectivity'].mean():.3f}")

# Sentimiento por rating
print(f"\nSentimiento promedio por rating:")
sentiment_by_rating = reviews_with_text.groupby('rating')['polarity'].mean().sort_index()
print(sentiment_by_rating)

# Visualizaci√≥n
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Distribuci√≥n de sentimientos
sentiment_counts = reviews_with_text['sentiment'].value_counts()
colors_sent = {'Positivo': '#4caf50', 'Neutral': '#ffc107', 'Negativo': '#f44336'}
axes[0, 0].bar(sentiment_counts.index, sentiment_counts.values,
               color=[colors_sent[s] for s in sentiment_counts.index], edgecolor='black', alpha=0.8)
axes[0, 0].set_ylabel('Cantidad de Reviews')
axes[0, 0].set_title('Distribuci√≥n de Sentimientos', fontweight='bold', fontsize=14)
axes[0, 0].grid(axis='y', alpha=0.3)
for i, v in enumerate(sentiment_counts.values):
    axes[0, 0].text(i, v + 50, str(v), ha='center', fontweight='bold', fontsize=12)

# Distribuci√≥n de polarity
axes[0, 1].hist(reviews_with_text['polarity'], bins=50, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 1].axvline(0, color='red', linestyle='--', linewidth=2, label='Neutral')
axes[0, 1].axvline(reviews_with_text['polarity'].mean(), color='green', linestyle='--', linewidth=2,
                   label=f'Media: {reviews_with_text["polarity"].mean():.3f}')
axes[0, 1].set_xlabel('Polarity (-1: Negativo, +1: Positivo)')
axes[0, 1].set_ylabel('Frecuencia')
axes[0, 1].set_title('Distribuci√≥n de Polarity (Sentimiento)', fontweight='bold')
axes[0, 1].legend()

# Polarity vs Rating
axes[1, 0].scatter(reviews_with_text['rating'], reviews_with_text['polarity'],
                   alpha=0.5, c=reviews_with_text['rating'], cmap='RdYlGn',
                   edgecolors='black', linewidth=0.5, s=30)
axes[1, 0].set_xlabel('Rating')
axes[1, 0].set_ylabel('Polarity')
axes[1, 0].set_title('Relaci√≥n: Rating vs Polarity', fontweight='bold')
axes[1, 0].axhline(0, color='red', linestyle='--', linewidth=1, alpha=0.5)
axes[1, 0].grid(alpha=0.3)

# Boxplot de polarity por rating
rating_groups = [reviews_with_text[reviews_with_text['rating'] == r]['polarity'].dropna()
                 for r in sorted(reviews_with_text['rating'].unique())]
bp = axes[1, 1].boxplot(rating_groups, labels=sorted(reviews_with_text['rating'].unique()),
                        patch_artist=True)
for patch, color in zip(bp['boxes'], ['#d32f2f', '#f57c00', '#fbc02d', '#7cb342', '#388e3c']):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[1, 1].set_xlabel('Rating')
axes[1, 1].set_ylabel('Polarity')
axes[1, 1].set_title('Distribuci√≥n de Polarity por Rating', fontweight='bold')
axes[1, 1].axhline(0, color='red', linestyle='--', linewidth=1, alpha=0.5)
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 7. AN√ÅLISIS DE PALABRAS FRECUENTES
# =========================================================
print("\n\nüî§ AN√ÅLISIS DE PALABRAS FRECUENTES")
print("-" * 80)

# Funci√≥n para limpiar texto
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

# Stopwords en espa√±ol e ingl√©s
stopwords_es = {'el', 'la', 'de', 'que', 'y', 'a', 'en', 'un', 'ser', 'se', 'no', 'haber', 'por', 'con', 'su', 'para',
                'como', 'estar', 'tener', 'le', 'lo', 'todo', 'pero', 'm√°s', 'hacer', 'o', 'poder', 'decir', 'este',
                'ir', 'otro', 'ese', 'la', 'si', 'me', 'ya', 'ver', 'porque', 'dar', 'cuando', '√©l', 'muy', 'sin',
                'vez', 'mucho', 'saber', 'qu√©', 'sobre', 'mi', 'alguno', 'mismo', 'yo', 'tambi√©n', 'hasta', 'a√±o',
                'dos', 'querer', 'entre', 'as√≠', 'primero', 'desde', 'grande', 'eso', 'ni', 'nos', 'llegar', 'pasar',
                'tiempo', 'ella', 's√≠', 'd√≠a', 'uno', 'bien', 'poco', 'deber', 'entonces', 'poner', 'cosa', 'tanto',
                'hombre', 'parecer', 'nuestro', 'tan', 'donde', 'ahora', 'parte', 'despu√©s', 'vida', 'quedar', 'siempre',
                'creer', 'hablar', 'llevar', 'dejar', 'nada', 'cada', 'seguir', 'menos', 'nuevo', 'encontrar', 'algo',
                'solo', 'decir', 'pueden', 'puede', 'the', 'is', 'are', 'was', 'were', 'has', 'have', 'had', 'be',
                'been', 'being', 'do', 'does', 'did', 'will', 'would', 'should', 'could', 'may', 'might', 'must',
                'can', 'this', 'that', 'these', 'those', 'i', 'you', 'he', 'she', 'it', 'we', 'they', 'them', 'their',
                'what', 'which', 'who', 'when', 'where', 'why', 'how', 'all', 'each', 'every', 'both', 'few', 'more',
                'most', 'other', 'some', 'such', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 'just', 'but',
                'and', 'or', 'if', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against',
                'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down',
                'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once'}

# Extraer palabras por sentimiento
def extract_words(df_subset):
    all_text = ' '.join(df_subset['text'].apply(clean_text))
    words = all_text.split()
    words = [w for w in words if len(w) > 3 and w not in stopwords_es]
    return Counter(words)

positive_words = extract_words(reviews_with_text[reviews_with_text['sentiment'] == 'Positivo'])
negative_words = extract_words(reviews_with_text[reviews_with_text['sentiment'] == 'Negativo'])
all_words = extract_words(reviews_with_text)

print(f"Top 20 palabras m√°s frecuentes (general):")
for word, count in all_words.most_common(20):
    print(f"  {word}: {count}")

print(f"\nTop 15 palabras en reviews positivas:")
for word, count in positive_words.most_common(15):
    print(f"  {word}: {count}")

print(f"\nTop 15 palabras en reviews negativas:")
for word, count in negative_words.most_common(15):
    print(f"  {word}: {count}")

# Visualizaci√≥n
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Top 20 palabras generales
top_words = dict(all_words.most_common(20))
axes[0, 0].barh(range(len(top_words)), list(top_words.values()),
                color=plt.cm.viridis(np.linspace(0, 1, len(top_words))))
axes[0, 0].set_yticks(range(len(top_words)))
axes[0, 0].set_yticklabels(list(top_words.keys()))
axes[0, 0].set_xlabel('Frecuencia')
axes[0, 0].set_title('Top 20 Palabras M√°s Frecuentes', fontweight='bold', fontsize=12)
axes[0, 0].invert_yaxis()

# Top 15 palabras positivas
top_pos = dict(positive_words.most_common(15))
axes[0, 1].barh(range(len(top_pos)), list(top_pos.values()), color='#4caf50', alpha=0.7)
axes[0, 1].set_yticks(range(len(top_pos)))
axes[0, 1].set_yticklabels(list(top_pos.keys()))
axes[0, 1].set_xlabel('Frecuencia')
axes[0, 1].set_title('Top 15 Palabras en Reviews Positivas', fontweight='bold', fontsize=12)
axes[0, 1].invert_yaxis()

# Top 15 palabras negativas
top_neg = dict(negative_words.most_common(15))
axes[1, 0].barh(range(len(top_neg)), list(top_neg.values()), color='#f44336', alpha=0.7)
axes[1, 0].set_yticks(range(len(top_neg)))
axes[1, 0].set_yticklabels(list(top_neg.keys()))
axes[1, 0].set_xlabel('Frecuencia')
axes[1, 0].set_title('Top 15 Palabras en Reviews Negativas', fontweight='bold', fontsize=12)
axes[1, 0].invert_yaxis()

# Subjectivity distribution
axes[1, 1].hist(reviews_with_text['subjectivity'], bins=50, color='mediumpurple', edgecolor='black', alpha=0.7)
axes[1, 1].axvline(reviews_with_text['subjectivity'].mean(), color='red', linestyle='--', linewidth=2,
                   label=f'Media: {reviews_with_text["subjectivity"].mean():.3f}')
axes[1, 1].set_xlabel('Subjectivity (0: Objetivo, 1: Subjetivo)')
axes[1, 1].set_ylabel('Frecuencia')
axes[1, 1].set_title('Distribuci√≥n de Subjetividad', fontweight='bold', fontsize=12)
axes[1, 1].legend()

plt.tight_layout()
plt.show()

# 8. WORD CLOUDS
# =========================================================
print("\n\n‚òÅÔ∏è GENERANDO WORD CLOUDS...")
print("-" * 80)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# WordCloud general
all_text = ' '.join(reviews_with_text['text'].apply(clean_text))
wordcloud_all = WordCloud(width=800, height=400, background_color='white',
                          stopwords=stopwords_es, colormap='viridis', max_words=100).generate(all_text)
axes[0].imshow(wordcloud_all, interpolation='bilinear')
axes[0].axis('off')
axes[0].set_title('Word Cloud - Todas las Reviews', fontweight='bold', fontsize=14)

# WordCloud positivas
positive_text = ' '.join(reviews_with_text[reviews_with_text['sentiment'] == 'Positivo']['text'].apply(clean_text))
if len(positive_text) > 0:
    wordcloud_pos = WordCloud(width=800, height=400, background_color='white',
                              stopwords=stopwords_es, colormap='Greens', max_words=100).generate(positive_text)
    axes[1].imshow(wordcloud_pos, interpolation='bilinear')
axes[1].axis('off')
axes[1].set_title('Word Cloud - Reviews Positivas', fontweight='bold', fontsize=14)

# WordCloud negativas
negative_text = ' '.join(reviews_with_text[reviews_with_text['sentiment'] == 'Negativo']['text'].apply(clean_text))
if len(negative_text) > 0:
    wordcloud_neg = WordCloud(width=800, height=400, background_color='white',
                              stopwords=stopwords_es, colormap='Reds', max_words=100).generate(negative_text)
    axes[2].imshow(wordcloud_neg, interpolation='bilinear')
axes[2].axis('off')
axes[2].set_title('Word Cloud - Reviews Negativas', fontweight='bold', fontsize=14)

plt.tight_layout()
plt.show()

# 9. AN√ÅLISIS POR CATEGOR√çA Y DISTRITO
# =========================================================
print("\n\nüèõÔ∏è AN√ÅLISIS DE REVIEWS POR CATEGOR√çA Y DISTRITO")
print("-" * 80)

# Reviews por categor√≠a
cat_review_stats = reviews_df.groupby('categoria').agg({
    'rating': ['mean', 'count', 'std'],
    'place_name': 'nunique'
}).round(2)
cat_review_stats.columns = ['rating_mean', 'review_count', 'rating_std', 'unique_places']
cat_review_stats = cat_review_stats[cat_review_stats['review_count'] >= 20].sort_values('rating_mean', ascending=False)

print(f"Top 10 categor√≠as por rating promedio (m√≠n. 20 reviews):")
print(cat_review_stats.head(10))

# Reviews por distrito
dist_review_stats = reviews_df.groupby('distrito').agg({
    'rating': ['mean', 'count', 'std'],
    'place_name': 'nunique'
}).round(2)
dist_review_stats.columns = ['rating_mean', 'review_count', 'rating_std', 'unique_places']
dist_review_stats = dist_review_stats[dist_review_stats['review_count'] >= 30].sort_values('rating_mean', ascending=False)

print(f"\nTop 10 distritos por rating promedio (m√≠n. 30 reviews):")
print(dist_review_stats.head(10))

# Visualizaci√≥n
fig, axes = plt.subplots(2, 2, figsize=(18, 14))

# Top categor√≠as por rating
top_cat = cat_review_stats.head(12)
axes[0, 0].barh(range(len(top_cat)), top_cat['rating_mean'], color='mediumseagreen', alpha=0.7)
axes[0, 0].set_yticks(range(len(top_cat)))
axes[0, 0].set_yticklabels(top_cat.index, fontsize=9)
axes[0, 0].set_xlabel('Rating Promedio')
axes[0, 0].set_title('Top 12 Categor√≠as por Rating (Reviews)', fontweight='bold')
axes[0, 0].invert_yaxis()
axes[0, 0].axvline(reviews_df['rating'].mean(), color='red', linestyle='--', linewidth=2)
for i, v in enumerate(top_cat['rating_mean']):
    axes[0, 0].text(v + 0.02, i, f'{v:.2f}', va='center', fontsize=9)

# Categor√≠as con m√°s reviews
cat_most_reviews = reviews_df.groupby('categoria').size().sort_values(ascending=False).head(12)
axes[0, 1].barh(range(len(cat_most_reviews)), cat_most_reviews.values, color='coral', alpha=0.7)
axes[0, 1].set_yticks(range(len(cat_most_reviews)))
axes[0, 1].set_yticklabels(cat_most_reviews.index, fontsize=9)
axes[0, 1].set_xlabel('N√∫mero de Reviews')
axes[0, 1].set_title('Top 12 Categor√≠as con M√°s Reviews', fontweight='bold')
axes[0, 1].invert_yaxis()
for i, v in enumerate(cat_most_reviews.values):
    axes[0, 1].text(v + 10, i, str(v), va='center', fontsize=9)

# Top distritos por rating
top_dist = dist_review_stats.head(12)
axes[1, 0].barh(range(len(top_dist)), top_dist['rating_mean'], color='steelblue', alpha=0.7)
axes[1, 0].set_yticks(range(len(top_dist)))
axes[1, 0].set_yticklabels(top_dist.index, fontsize=9)
axes[1, 0].set_xlabel('Rating Promedio')
axes[1, 0].set_title('Top 12 Distritos por Rating (Reviews)', fontweight='bold')
axes[1, 0].invert_yaxis()
axes[1, 0].axvline(reviews_df['rating'].mean(), color='red', linestyle='--', linewidth=2)
for i, v in enumerate(top_dist['rating_mean']):
    axes[1, 0].text(v + 0.02, i, f'{v:.2f}', va='center', fontsize=9)

# Distritos con m√°s reviews
dist_most_reviews = reviews_df.groupby('distrito').size().sort_values(ascending=False).head(12)
axes[1, 1].barh(range(len(dist_most_reviews)), dist_most_reviews.values, color='lightgreen', alpha=0.7)
axes[1, 1].set_yticks(range(len(dist_most_reviews)))
axes[1, 1].set_yticklabels(dist_most_reviews.index, fontsize=9)
axes[1, 1].set_xlabel('N√∫mero de Reviews')
axes[1, 1].set_title('Top 12 Distritos con M√°s Reviews', fontweight='bold')
axes[1, 1].invert_yaxis()
for i, v in enumerate(dist_most_reviews.values):
    axes[1, 1].text(v + 10, i, str(v), va='center', fontsize=9)

plt.tight_layout()
plt.show()

# 10. AN√ÅLISIS TEMPORAL DE REVIEWS
# =========================================================
print("\n\n‚è∞ AN√ÅLISIS TEMPORAL DE REVIEWS")
print("-" * 80)

# Extraer informaci√≥n temporal
def parse_time(time_str):
    """Convierte 'X months ago', 'X years ago', etc. a meses"""
    if pd.isna(time_str) or time_str == '':
        return None

    time_str = str(time_str).lower()

    if 'month' in time_str:
        try:
            months = int(re.findall(r'\d+', time_str)[0])
            return months
        except:
            return None
    elif 'year' in time_str:
        try:
            years = int(re.findall(r'\d+', time_str)[0])
            return years * 12
        except:
            return None
    elif 'week' in time_str:
        try:
            weeks = int(re.findall(r'\d+', time_str)[0])
            return weeks / 4
        except:
            return None
    elif 'day' in time_str:
        return 0
    else:
        return None

reviews_df['months_ago'] = reviews_df['time'].apply(parse_time)
reviews_temporal = reviews_df[reviews_df['months_ago'].notna()].copy()

print(f"Reviews con informaci√≥n temporal: {len(reviews_temporal)}")
if len(reviews_temporal) > 0:
    print(f"Review m√°s antigua: {reviews_temporal['months_ago'].max():.0f} meses atr√°s")
    print(f"Review m√°s reciente: {reviews_temporal['months_ago'].min():.0f} meses atr√°s")
    print(f"Promedio de antig√ºedad: {reviews_temporal['months_ago'].mean():.1f} meses")

# Visualizaci√≥n temporal
if len(reviews_temporal) > 0:
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # Distribuci√≥n temporal
    axes[0, 0].hist(reviews_temporal['months_ago'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    axes[0, 0].set_xlabel('Meses Atr√°s')
    axes[0, 0].set_ylabel('N√∫mero de Reviews')
    axes[0, 0].set_title('Distribuci√≥n Temporal de Reviews', fontweight='bold')
    axes[0, 0].grid(axis='y', alpha=0.3)

    # Reviews por per√≠odo
    reviews_temporal['period'] = pd.cut(reviews_temporal['months_ago'],
                                       bins=[0, 3, 6, 12, 24, 100],
                                       labels=['0-3 meses', '3-6 meses', '6-12 meses', '1-2 a√±os', '2+ a√±os'])
    period_counts = reviews_temporal['period'].value_counts().sort_index()
    axes[0, 1].bar(range(len(period_counts)), period_counts.values,
                   color=['#4caf50', '#8bc34a', '#ffeb3b', '#ff9800', '#f44336'],
                   edgecolor='black', alpha=0.7)
    axes[0, 1].set_xticks(range(len(period_counts)))
    axes[0, 1].set_xticklabels(period_counts.index, rotation=45, ha='right')
    axes[0, 1].set_ylabel('N√∫mero de Reviews')
    axes[0, 1].set_title('Reviews por Per√≠odo de Tiempo', fontweight='bold')
    axes[0, 1].grid(axis='y', alpha=0.3)

    # Rating por per√≠odo
    rating_by_period = reviews_temporal.groupby('period')['rating'].mean().sort_index()
    axes[1, 0].plot(range(len(rating_by_period)), rating_by_period.values,
                    marker='o', linewidth=3, markersize=10, color='#2196f3')
    axes[1, 0].set_xticks(range(len(rating_by_period)))
    axes[1, 0].set_xticklabels(rating_by_period.index, rotation=45, ha='right')
    axes[1, 0].set_ylabel('Rating Promedio')
    axes[1, 0].set_title('Evoluci√≥n del Rating Promedio por Per√≠odo', fontweight='bold')
    axes[1, 0].axhline(reviews_temporal['rating'].mean(), color='red', linestyle='--',
                      linewidth=2, label=f'Media: {reviews_temporal["rating"].mean():.2f}')
    axes[1, 0].legend()
    axes[1, 0].grid(alpha=0.3)

    # Sentimiento por per√≠odo (si est√° disponible)
    if 'sentiment' in reviews_temporal.columns:
        sentiment_period = pd.crosstab(reviews_temporal['period'], reviews_temporal['sentiment'], normalize='index') * 100
        sentiment_period = sentiment_period.reindex(['0-3 meses', '3-6 meses', '6-12 meses', '1-2 a√±os', '2+ a√±os'])
        sentiment_period.plot(kind='bar', stacked=True, ax=axes[1, 1],
                            color={'Positivo': '#4caf50', 'Neutral': '#ffc107', 'Negativo': '#f44336'},
                            edgecolor='black', alpha=0.8)
        axes[1, 1].set_xlabel('Per√≠odo')
        axes[1, 1].set_ylabel('Porcentaje (%)')
        axes[1, 1].set_title('Distribuci√≥n de Sentimientos por Per√≠odo', fontweight='bold')
        axes[1, 1].legend(title='Sentimiento')
        axes[1, 1].set_xticklabels(sentiment_period.index, rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

# 11. AN√ÅLISIS DE REVIEWS EXTREMAS
# =========================================================
print("\n\n‚ö†Ô∏è AN√ÅLISIS DE REVIEWS EXTREMAS")
print("-" * 80)

# Reviews 5 estrellas
five_star = reviews_df[reviews_df['rating'] == 5]
print(f"Reviews con 5 estrellas: {len(five_star)} ({len(five_star)/len(reviews_df)*100:.1f}%)")

# Reviews 1 estrella
one_star = reviews_df[reviews_df['rating'] == 1]
print(f"Reviews con 1 estrella: {len(one_star)} ({len(one_star)/len(reviews_df)*100:.1f}%)")

# Lugares con m√°s reviews 5 estrellas
if len(five_star) > 0:
    top_five_star = five_star['place_name'].value_counts().head(10)
    print(f"\nTop 10 lugares con m√°s reviews de 5 estrellas:")
    print(top_five_star)

# Lugares con m√°s reviews 1 estrella
if len(one_star) > 0:
    top_one_star = one_star['place_name'].value_counts().head(10)
    print(f"\nTop 10 lugares con m√°s reviews de 1 estrella:")
    print(top_one_star)

# Visualizaci√≥n
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Distribuci√≥n por rating
rating_dist = reviews_df['rating'].value_counts().sort_index()
colors_rating = ['#d32f2f', '#f57c00', '#fbc02d', '#7cb342', '#388e3c']
axes[0, 0].bar(rating_dist.index, rating_dist.values, color=colors_rating, edgecolor='black', alpha=0.8)
axes[0, 0].set_xlabel('Rating')
axes[0, 0].set_ylabel('N√∫mero de Reviews')
axes[0, 0].set_title('Distribuci√≥n Completa de Ratings', fontweight='bold', fontsize=14)
axes[0, 0].grid(axis='y', alpha=0.3)
for i, (rating, count) in enumerate(zip(rating_dist.index, rating_dist.values)):
    axes[0, 0].text(rating, count + 50, str(count), ha='center', fontweight='bold')

# Top lugares 5 estrellas
if len(five_star) > 0:
    axes[0, 1].barh(range(len(top_five_star)), top_five_star.values, color='#388e3c', alpha=0.7)
    axes[0, 1].set_yticks(range(len(top_five_star)))
    axes[0, 1].set_yticklabels(top_five_star.index, fontsize=8)
    axes[0, 1].set_xlabel('N√∫mero de Reviews 5‚òÖ')
    axes[0, 1].set_title('Lugares con M√°s Reviews de 5 Estrellas', fontweight='bold')
    axes[0, 1].invert_yaxis()

# Top lugares 1 estrella
if len(one_star) > 0:
    axes[1, 0].barh(range(len(top_one_star)), top_one_star.values, color='#d32f2f', alpha=0.7)
    axes[1, 0].set_yticks(range(len(top_one_star)))
    axes[1, 0].set_yticklabels(top_one_star.index, fontsize=8)
    axes[1, 0].set_xlabel('N√∫mero de Reviews 1‚òÖ')
    axes[1, 0].set_title('Lugares con M√°s Reviews de 1 Estrella', fontweight='bold')
    axes[1, 0].invert_yaxis()

# Categor√≠as con m√°s reviews extremas
extreme_reviews = pd.concat([
    reviews_df[reviews_df['rating'] == 5].groupby('categoria').size().rename('5_stars'),
    reviews_df[reviews_df['rating'] == 1].groupby('categoria').size().rename('1_star')
], axis=1).fillna(0)
extreme_reviews['total'] = extreme_reviews['5_stars'] + extreme_reviews['1_star']
extreme_reviews = extreme_reviews.sort_values('total', ascending=False).head(10)

x = range(len(extreme_reviews))
width = 0.35
axes[1, 1].bar([i - width/2 for i in x], extreme_reviews['5_stars'], width,
               label='5 Estrellas', color='#388e3c', alpha=0.8)
axes[1, 1].bar([i + width/2 for i in x], extreme_reviews['1_star'], width,
               label='1 Estrella', color='#d32f2f', alpha=0.8)
axes[1, 1].set_xticks(x)
axes[1, 1].set_xticklabels(extreme_reviews.index, rotation=45, ha='right', fontsize=8)
axes[1, 1].set_ylabel('N√∫mero de Reviews')
axes[1, 1].set_title('Reviews Extremas por Categor√≠a (Top 10)', fontweight='bold')
axes[1, 1].legend()
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 12. CORRELACIONES
# =========================================================
print("\n\nüìä AN√ÅLISIS DE CORRELACIONES")
print("-" * 80)

# Crear DataFrame con m√©tricas agregadas por lugar
place_metrics = df.groupby('name').agg({
    'rating': 'first',
    'num_reviews': 'first',
    'categoria': 'first',
    'distrito': 'first'
}).reset_index()

# Calcular m√©tricas adicionales de reviews
review_metrics = reviews_df.groupby('place_name').agg({
    'rating': ['mean', 'std', 'count']
}).reset_index()
review_metrics.columns = ['name', 'review_rating_mean', 'review_rating_std', 'review_count']

place_metrics = place_metrics.merge(review_metrics, on='name', how='left')

# Agregar m√©tricas de sentimiento si existen
if 'sentiment' in reviews_with_text.columns:
    sentiment_metrics = reviews_with_text.groupby('place_name').agg({
        'polarity': 'mean',
        'subjectivity': 'mean',
        'text_length': 'mean',
        'word_count': 'mean'
    }).reset_index()
    sentiment_metrics.columns = ['name', 'avg_polarity', 'avg_subjectivity', 'avg_text_length', 'avg_word_count']
    place_metrics = place_metrics.merge(sentiment_metrics, on='name', how='left')

# Matriz de correlaci√≥n
numeric_cols = place_metrics.select_dtypes(include=[np.number]).columns
corr_matrix = place_metrics[numeric_cols].corr()

print("Matriz de correlaci√≥n:")
print(corr_matrix)

# Visualizaci√≥n
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Heatmap de correlaci√≥n
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', square=True, ax=axes[0], cbar_kws={'label': 'Correlaci√≥n'})
axes[0].set_title('Matriz de Correlaci√≥n de M√©tricas', fontweight='bold', fontsize=14)

# Correlaciones con rating
rating_corr = corr_matrix['rating'].sort_values(ascending=False)[1:]  # Excluir correlaci√≥n consigo mismo
colors = ['#4caf50' if x > 0 else '#f44336' for x in rating_corr.values]
axes[1].barh(range(len(rating_corr)), rating_corr.values, color=colors, alpha=0.7, edgecolor='black')
axes[1].set_yticks(range(len(rating_corr)))
axes[1].set_yticklabels(rating_corr.index)
axes[1].set_xlabel('Correlaci√≥n con Rating')
axes[1].set_title('Correlaciones con el Rating del Lugar', fontweight='bold', fontsize=14)
axes[1].axvline(0, color='black', linestyle='-', linewidth=1)
axes[1].invert_yaxis()
for i, v in enumerate(rating_corr.values):
    axes[1].text(v + 0.01 if v > 0 else v - 0.01, i, f'{v:.3f}',
                va='center', ha='left' if v > 0 else 'right', fontsize=9)

plt.tight_layout()
plt.show()

# 13. RESUMEN EJECUTIVO DE REVIEWS
# =========================================================
print("\n\n" + "=" * 80)
print("üìã RESUMEN EJECUTIVO - AN√ÅLISIS DE REVIEWS")
print("=" * 80)

total_reviews = len(reviews_df)
reviews_with_text_count = reviews_with_text.shape[0] if 'reviews_with_text' in locals() else 0
avg_rating = reviews_df['rating'].mean()
places_with_reviews = reviews_df['place_name'].nunique()

print(f"""
VOLUMEN DE REVIEWS:
- Total de reviews: {total_reviews:,}
- Lugares con reviews: {places_with_reviews}
- Reviews por lugar (promedio): {total_reviews/places_with_reviews:.1f}
- Reviews con texto: {reviews_with_text_count:,} ({reviews_with_text_count/total_reviews*100:.1f}%)

RATINGS:
- Rating promedio: {avg_rating:.2f}/5.0
- Rating m√°s com√∫n: {reviews_df['rating'].mode().values[0]:.0f} estrellas
- Reviews 5‚òÖ: {len(five_star)} ({len(five_star)/total_reviews*100:.1f}%)
- Reviews 1‚òÖ: {len(one_star)} ({len(one_star)/total_reviews*100:.1f}%)
""")

if 'sentiment' in reviews_with_text.columns:
    pos_pct = (reviews_with_text['sentiment'] == 'Positivo').sum() / len(reviews_with_text) * 100
    neg_pct = (reviews_with_text['sentiment'] == 'Negativo').sum() / len(reviews_with_text) * 100
    neu_pct = (reviews_with_text['sentiment'] == 'Neutral').sum() / len(reviews_with_text) * 100

    print(f"""AN√ÅLISIS DE SENTIMIENTO:
- Sentimiento positivo: {pos_pct:.1f}%
- Sentimiento neutral: {neu_pct:.1f}%
- Sentimiento negativo: {neg_pct:.1f}%
- Polarity promedio: {reviews_with_text['polarity'].mean():.3f}
- Subjetividad promedio: {reviews_with_text['subjectivity'].mean():.3f}
""")

print(f"""CARACTER√çSTICAS DE TEXTO:
- Longitud promedio: {reviews_with_text['text_length'].mean():.0f} caracteres
- Palabras promedio: {reviews_with_text['word_count'].mean():.1f} palabras
- Idioma principal: {reviews_with_text['language'].mode().values[0] if len(reviews_with_text) > 0 else 'N/A'}
""")

print(f"""TOP 3 CATEGOR√çAS CON M√ÅS REVIEWS:
{chr(10).join([f"  {i+1}. {cat}: {count} reviews" for i, (cat, count) in enumerate(reviews_df.groupby('categoria').size().sort_values(ascending=False).head(3).items())])}

TOP 3 LUGARES CON M√ÅS REVIEWS:
{chr(10).join([f"  {i+1}. {place}: {count} reviews" for i, (place, count) in enumerate(reviews_df['place_name'].value_counts().head(3).items())])}
""")

if len(positive_words) > 0:
    print(f"""TOP 5 PALABRAS EN REVIEWS POSITIVAS:
{chr(10).join([f"  {i+1}. {word} ({count} veces)" for i, (word, count) in enumerate(positive_words.most_common(5))])}
""")

if len(negative_words) > 0:
    print(f"""TOP 5 PALABRAS EN REVIEWS NEGATIVAS:
{chr(10).join([f"  {i+1}. {word} ({count} veces)" for i, (word, count) in enumerate(negative_words.most_common(5))])}
""")

print("=" * 80)
print("‚úÖ AN√ÅLISIS COMPLETADO")
print("=" * 80)
```

```python
import pandas as pd
import ast

# Cargar tu CSV
data = pd.read_csv("google.csv")

# Convertir la columna reviews (string) a lista de diccionarios
data["reviews"] = data["reviews"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

# Crear lista de filas desanidadas
rows = []
for idx, row in data.iterrows():
    place_id = idx  # o podr√≠as usar un campo √∫nico de 'name' o 'url'
    for review in row["reviews"]:
        rows.append({
            "place_id": place_id,
            "user_id": review.get("author"),
            "rating_individual": review.get("rating"),
            "review_text": review.get("text"),
            "time": review.get("time")
        })

# Crear nuevo DataFrame con las rese√±as individuales
reviews_df = pd.DataFrame(rows)

# Agregar ID √∫nico
reviews_df["review_id"] = range(1, len(reviews_df) + 1)

# Reordenar columnas
reviews_df = reviews_df[["review_id", "place_id", "user_id", "rating_individual", "review_text", "time"]]

# Guardar CSV
reviews_df.to_csv("google_reviews_flattened.csv", index=False, encoding="utf-8-sig")

print("‚úÖ Rese√±as desanidadas correctamente:", reviews_df.shape)
```
